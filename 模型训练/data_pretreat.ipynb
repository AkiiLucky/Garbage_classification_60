```python
import pandas as pd
import numpy as np
from PIL import Image
import cv2
import os
import torch
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
plt.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

```python
labelIndex = {0:'可回收垃圾',1:'有害垃圾',2:'厨余垃圾',3:'其他垃圾'} #一级分类的索引
```

```python
indexList = []
for i in range(4):
    index = pd.read_csv(f".//垃圾目录//{i} {labelIndex[i]}.txt")
    indexList.append(index)
for i in range(len(indexList)):
    indexList[i][labelIndex[i]] = indexList[i][labelIndex[i]].map(lambda x: x.replace(f'{i}\t',''))
# print(indexList)  
```

```python
count = 0
index_class60 = {}
for i in range(4):
    for j in range(len(indexList[i])):
        name = indexList[i][labelIndex[i]][j]
        index_class60[name] = count
        count += 1
print(index_class60) # 二级分类的索引
```

```python
dataset = []
def read_directory(directory_name, y):
    for filename in os.listdir(directory_name):
        img = cv2.imdecode(np.fromfile(os.path.join(directory_name, filename), dtype=np.uint8), cv2.IMREAD_COLOR)
        #img.resize(224,224,3)
        if img is None:            
            print(f"import {directory_name}/{filename} failed")
            continue
        img = cv2.resize(img, (224, 224), interpolation=cv2.INTER_CUBIC)
        x = torch.from_numpy(img)
        dataset.append([x,y])
        print(f"import {directory_name}/{filename} success","标签:",y)
        # if (len(dataset)>10000):
        #     break
```
